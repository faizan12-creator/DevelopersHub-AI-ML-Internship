```python
import gradio as gr
import requests
import warnings
warnings.filterwarnings('ignore')

FALLBACK = {
    'fever'         : "Fever is usually caused by infection. Drink fluids, rest, and take paracetamol. See a doctor if temperature exceeds 103°F.",
    'headache'      : "Headaches can result from stress, dehydration, or lack of sleep. Drink water and rest. See a doctor if severe.",
    'cough'         : "Cough may be due to cold, allergies or infection. Stay hydrated. See a doctor if it persists beyond 2 weeks.",
    'cold'          : "For common cold: rest, stay hydrated, take vitamin C. Usually resolves in 7-10 days.",
    'paracetamol'   : "Paracetamol is generally safe for children at correct weight-based doses. Always consult your doctor first.",
    'diabetes'      : "Diabetes symptoms include frequent urination, excessive thirst and fatigue. Manage with diet, exercise and medication.",
    'blood pressure': "To reduce high blood pressure: reduce salt intake, exercise regularly and avoid stress. Take prescribed medications.",
    'throat'        : "Sore throat is usually viral. Gargle warm salt water, stay hydrated. See doctor if very severe.",
    'stomach'       : "Stomach pain may be from indigestion or infection. Rest and drink water. See doctor if pain is severe.",
    'heart'         : "Heart disease symptoms: chest pain, breathlessness. This is serious — see a doctor immediately!"
}

HARMFUL = ['suicide', 'kill', 'overdose', 'self harm', 'poison']
HEALTH  = ['pain', 'fever', 'headache', 'cough', 'cold', 'medicine',
           'doctor', 'symptom', 'disease', 'health', 'sick', 'throat',
           'stomach', 'heart', 'blood', 'pressure', 'diabetes',
           'paracetamol', 'tablet', 'hospital', 'infection']

def chatbot(message, history):
    msg = message.lower()
    
    if any(w in msg for w in HARMFUL):
        return "⚠️ I cannot help with this. Please call Emergency: 115 immediately."
    
    if not any(w in msg for w in HEALTH):
        return "🤖 I only answer health questions!\n\nTry: 'What causes fever?' or 'Is paracetamol safe?'"
    
    answer = next((v for k, v in FALLBACK.items() if k in msg),
                  "Please consult a qualified doctor for accurate advice. 🏥")
    
    return answer + "\n\n⚕️ *Always consult a doctor for medical advice.*"

print("✅ Chatbot ready!")
```

```python
demo = gr.ChatInterface(
    fn=chatbot,
    title="🏥 Health Query Chatbot",
    description="### DevelopersHub Corporation | AI/ML Internship — Task 4\nAsk me any general health question!",
    examples=[
        "What causes a sore throat?",
        "Is paracetamol safe for children?",
        "What are symptoms of diabetes?",
        "How to reduce high blood pressure?",
        "What causes headaches?"
    ]
)

demo.launch(share=True)
```
